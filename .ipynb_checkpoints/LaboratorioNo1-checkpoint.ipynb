## ¿Que es GIT?

Git es un sistema de versíon de controles, los sitemas de conrol de versiones son un tipo de software que perimte realizar un seguimiento de los cambios efectuados en el código a lo largo del tiempo. A medida que el desarrollador edita el código, el sistema de control de versiónes toma un snapshot de los archivos y posteriormente guarda este snapshot de forma permanete para que se pueda recuperar más adelante si fuese requerido. (Mijacobs, 2023)

Si no se posee un control de versiónes, los dessarrolladores se sienten tentados a mantener varias copias del codigo en su equipo. Esto puede representar un problema, debido a que es fácil cambiar o eliminar un archivo en la copia incorrecta del código, lo que podría ocasionar que el dessarrollador pierda parte de su trabajo. El control de versiónes permite administrar todas las versiones del código, pero presentan al equipo una única versión a la vez. (Mijacobs, 2023)

GIT es el estandar mundial para el control de versiones, al ser un control de versiones distribuido, este almacena una copia local del repositorio de control de versiones completo. Estos repositorios locales permiten trabajar sin conexión o de forma remora con facilidad. Los desarrolladores realizan el commit su trabajo de forma local y posteriomrente sincronizan la copia local con la del servidor. (Mijacobs, 2023)

### Aspectos basicos de Git 

Cada vez que se guarda el proyecto Git genera commit. Este commit es el snapshot de todos los archivos. En caso de que un archivo no haya cambiado de un commit al siguiente, Git utiliza el archivo previamente almacenado. (Microsoft Learn, 2023)

![](https://learn.microsoft.com/es-es/devops/_img/linear_straight_line.png)

Los commits crean vinculos con otros commits y forman un gráfico del historial del desarrollo. Es posible revertir alas versiones pasadas del proyecto e inspecciónas los cambios realizados entre las versiones. Los cambios son identificados en Git por un hash ciptográfico único del contenido de la confirmación, esto impide que se realicen cambios o se dañe la información sin que sea detectado por Git. (Microsoft Learn, 2023)

### Ramas 

Cada desarrollador guarda los cambios en su propio repositorio local, como resultado pueden existir diferentes cambios basados en el mismo commit. GIT provee herramientas para aislar los cambios y unificarlos posteriormente. Las ramas son los punteros del trabajo en curso que administran dicha separación. Una vez se commplete el trabajo creado en una rama se puee combinar de nuevo con la rama trocal del equipo.(Microsoft Learn, 2023)

![Branches](https://learn.microsoft.com/es-es/devops/_img/branching_line.png)

### Archivos y commits 

Los archivos en Git pueden poseer 3 estados:

- Modificados 
- Almacenados provisionalmente 
- Confrimados 

Cuando un archivo es modificado por primera vez los cambios solo existen en el directorio de trabajo. Aun no forman parte de un commit o del historial de desarrollo. Los desarrolladores deben almacenar provisonalmente los archivos modificados que se incluirán dentro del commit. El area de almacenamiento ptovisional contiene los cambios que se incluirán en el siguiente commit. Una vez el desarrollador esté satisfecho con los cambios almacenados provisionalmene, se empaquetan los archivos como un commit/confirmación con un mensaje que describe los cambios realizados y pasa a formar parte del historial de desarrrollo. (Microsoft Learn, 2023)

![estado de archivos](https://learn.microsoft.com/es-es/devops/_img/file_status_lifecycle.2.png)



## ¿Qué es GitHub?
GitHub 3s una plataforma basada en nube que permite almacenar, compartir y trabajar con otros colaboradoes al momento de escribir código. (GitHub Docs, S.F.)

Almacenar el código en un repositorio de GitHub permite lo siguiente:

- Compartir y presentar el proyecto 
- LLevar un regisrtro y control de cambios del código en el tiempo 
- permite que los otros colaboradores revisen el comdigo y realicen sugerencias 
- Collaborar en un proyecto compartido,  sin preocuparse de que los cambios de un individuo impacten el trabajo de los otros colaboradores previo a que este listo para integrarse. 

EL trabajo colaborativo es una de las funcines fundamentales de Github y esto es posible gracias al estandar de codigo abierto GIT, sobre el cual GitHub fue construido. (GitHub Docs, S.F.)

### ¿Cómo se integra Git con GitHub?

Cuando se cargan archivos en GitHub, estos son almacenados en un repositorio Git, esto significa que cuando se realizan cambio en los archivos de GitHub, Git automaticamente maneja y lleva el registro de los cambios.(GitHub Docs, S.F.)

Dentro de GitHub se pueden encontrar diversas acciones relacionadas a Git como crear un repositorio Git, crear ramas, actualizar o editar archivos.(GitHub Docs, S.F.)

La mayotia de personas trabajan con los arhcivos localmente en su equipo y posterormente sincronizan los cambios locales y todos los datos relacionados a Git con el servidor centraldel repositorio en GitHub. Una de las herramientas que se puede útilizar para realizar esto es GitHub Desktop.(GitHub Docs, S.F.)

Al inciar el trabajo colaborativo, donde multipes personas necesitan realizar cambios en el mismo repositorio al mismo tiempo se deben ejecutar 2 acciones continuamente:

- Pull: Esta accion recupera todos los cambios recientes confirmados en el repositorio de GitHub

- Push: Esta accion publica los cambios recientes hacia el repositorio de GitHUb.











 ## Markdown
 
Es un lenguaje marcado ligero que permite agregar formato a documentos de texto plano. 

Utilizar makdown es diferente a utilizar un editor WYSIWYG. En aplicaciones como microsoft word se utilizan diversos botones para agregar formato a palabras y parrafos y estos tienen un efecto visual inmediato. En cambio en Markdown se utiliza una sintaxis para indicar dentro del texto que palabras o frases deben verse diferentes. Por ejemplo, para denotar un título se debe agregar el simbolo (#) antes de el texto, o para colocar un texto en negrilla se debe colocar el textro entre 2 pares de (*). (Markdown,s.f.)

### ¿Como funcióna?
 Cuando se escribe en markdown, el texto es almacenado como texto plano en un archivo con extensión .md o .markdown. Para poder visualizar el archivo markdown con el formato aplicado se debe tener una aplicación capaz deprocesar este archivo. Estas aplicaciónes utilizan algo llamado el procesador de Markdown para convertir el texto en formato Markdoen en un formato de salida HTMl. En este punto el documento puede ser visualizado en un naveggador web. A continuación se ilustra una representación de este suceso.(Markdown,s.f.)
 
 
 
 Este proceso de 4 pasos se puede resumir de la siguiente forma:
 1. Crear un archivo markdown con un editor de texto o una applicacion dedicada de Markdown. El archivo debe tener una extensión .md o .markdown
 2. Abrir el archivo en una aplicación markdown 
 3. Usar la aplicación Markdown para converit el el archivo markdown en un documento HTML 
 4. Ver el arvhivo HTML en un navegador o utilizar la aplicación markdwon para convertir el archivo en otro formato como PDF
